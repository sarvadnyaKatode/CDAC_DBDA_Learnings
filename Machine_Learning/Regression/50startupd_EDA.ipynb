## Drop Administration Column

### Subtask:
Drop the 'Administration' column from the `df` DataFrame.


**Reasoning**:
To drop the 'Administration' column from the DataFrame, I will use the `df.drop()` method with `columns='Administration'` and `inplace=True`. Then, I will display the first 5 rows of the modified DataFrame using `df.head()` to verify the removal of the column.



```python
df.drop(columns=['Administration'], inplace=True)
print("DataFrame after dropping 'Administration' column:")
print(df.head())
```

**Reasoning**:
The previous subtask of dropping the 'Administration' column is complete. The next step is to redefine the feature matrix `X` and target vector `y` based on the modified `df` DataFrame. `X` should contain all columns except 'Profit', and `y` should contain only the 'Profit' column.



```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

print("Shape of X after dropping 'Administration' column:", X.shape)
print("Shape of y:", y.shape)
```

**Reasoning**:
To reapply one-hot encoding to the 'State' column, I will use `ColumnTransformer` with `OneHotEncoder`. The 'State' column is now at index 2 in the updated `X`.



```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer([('ohe', OneHotEncoder(), [2])],
                       remainder='passthrough')
X_trans = ct.fit_transform(X)

print("Shape of X_trans after re-applying one-hot encoding:", X_trans.shape)
print("First 5 rows of X_trans:\n", X_trans[:5])
```

**Reasoning**:
Since missing values in the original numerical columns ('RDSpend', 'MarketingSpend') were already imputed in `df` before `X` and `y` were redefined, there are no further missing values to impute in `X_trans`. The next step is to re-split the preprocessed data into training and testing sets.



```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_trans, y, test_size=0.2, random_state=0)

print("Shapes after re-splitting data:")
print(f"X_train: {X_train.shape}")
print(f"X_test: {X_test.shape}")
print(f"y_train: {y_train.shape}")
print(f"y_test: {y_test.shape}")
```

**Reasoning**:
Now that the data has been re-split after dropping the 'Administration' column, I need to retrain the Linear Regression model with the new `X_train` and `y_train` and make predictions on `X_test`.



```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Retrain Linear Regression model
linear_regressor_new = LinearRegression()
linear_regressor_new.fit(X_train, y_train)

# Make predictions
yhat_lr_new = linear_regressor_new.predict(X_test)

# Evaluate Linear Regression model
mse_lr_new = mean_squared_error(y_test, yhat_lr_new)
rmse_lr_new = np.sqrt(mse_lr_new)
r2_lr_new = r2_score(y_test, yhat_lr_new)

# Calculate Adjusted R-squared
n = len(y_test) # Number of observations
p = X_test.shape[1] # Number of features
adjusted_r2_lr_new = 1 - (1 - r2_lr_new) * (n - 1) / (n - p - 1)

print("Linear Regression Model (After dropping 'Administration'):")
print(f"  Mean Squared Error (MSE): {mse_lr_new:.2f}")
print(f"  Root Mean Squared Error (RMSE): {rmse_lr_new:.2f}")
print(f"  R-squared (R2): {r2_lr_new:.4f}")
print(f"  Adjusted R-squared: {adjusted_r2_lr_new:.4f}")
```

**Reasoning**:
After retraining and evaluating the Linear Regression model, the next step is to retrain and re-evaluate the Gradient Boosting Regressor model using the newly prepared `X_train` and `y_train` data, and then make predictions on `X_test`.



```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Retrain Gradient Boosting Regressor model
gbr_new = GradientBoostingRegressor(random_state=0)
gbr_new.fit(X_train, y_train)

# Make predictions
yhat_gb_new = gbr_new.predict(X_test)

# Evaluate Gradient Boosting Regressor model
mse_gb_new = mean_squared_error(y_test, yhat_gb_new)
rmse_gb_new = np.sqrt(mse_gb_new)
r2_gb_new = r2_score(y_test, yhat_gb_new)

# Calculate Adjusted R-squared
n = len(y_test) # Number of observations
p = X_test.shape[1] # Number of features
adjusted_r2_gb_new = 1 - (1 - r2_gb_new) * (n - 1) / (n - p - 1)

print("Gradient Boosting Regressor Model (After dropping 'Administration'):")
print(f"  Mean Squared Error (MSE): {mse_gb_new:.2f}")
print(f"  Root Mean Squared Error (RMSE): {rmse_gb_new:.2f}")
print(f"  R-squared (R2): {r2_gb_new:.4f}")
print(f"  Adjusted R-squared: {adjusted_r2_gb_new:.4f}")
```
